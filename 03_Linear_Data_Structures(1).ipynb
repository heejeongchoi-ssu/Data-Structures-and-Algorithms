# **1. 배열**
## 1-1. 연습 문제
[문제명] 나누어 떨어지는 숫자 배열
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/12910

```python
def solution(arr, divisor):
    answer = [i for i in arr if i % divisor == 0]
    return sorted(answer) if answer else [-1]
```

```python
print(solution(arr=[5, 9, 7, 10], divisor=5))  # 실행결과: [5, 10]
print(solution(arr=[2, 36, 1, 3], divisor=1))  # 실행결과: [1, 2, 3, 36]
print(solution(arr=[3, 2, 6], divisor=10))  # 실행결과: [-1]
```

# **2. 연결 리스트**
## 2-1. 단순 연결 리스트 구현

```python
class Node:
    def __init__(self, elem, next=None):
        self.data = elem                      # data는 값을 가리키는 변수(속성, attribute)
        self.next = next                      # next는 다음 노드를 가리키는 변수

    def append(self, node):                   # 현재 노드(self) 다음에 node를 넣는 연산
        if node is not None:                  # node가 None이 아니면
            node.next = self.next             # ① node의 next에 self 다음 노드를 연결
            self.next = node                  # ② 현재 노드(self)의 다음 노드는 node가 됨

    def popNext(self):                        # 현재 노드(self)의 다음 노드를 삭제하는 연산
        node = self.next                      # 현재 노드(self)의 다음 노드
        if node is not None:                  # node가 None이 아니면
            self.next = node.next             # self의 다음 노드는 node.next
        return node                           # 삭제할 다음 노드를 반환
```

```python
class LinkedList:                             # 단순 연결 리스트 클래스
    def __init__(self):                       # 생성자
        self.head = None                      # head 선언 및 None으로 초기화
        self.length = 0                       # length 선언 및 0으로 초기화

    def __len__(self):
        return self.length                    # len() 함수 사용시 length 반환

    def isEmpty(self):                        # 공백상태 검사
        return self.head is None              # head가 None이면 공백

    def isFull(self):                         # 포화상태 검사
        return False                          # 연결 구조에서는 포화상태 없음

    def getNode(self, pos):
        if pos < 0 or pos >= self.length:     # 잘못된 위치인 경우
            return None                       # None 반환
        ptr = self.head                       # 시작 위치 -> head
        for _ in range(pos):                  # pos번 링크를 따라 이동
            ptr = ptr.next                    # ptr을 다음 노드로 업데이트
        return ptr                            # 최종 노드(pos번째 노드)를 반환

    def getEntry(self, pos):
        node = self.getNode(pos)              # pos번째 노드를 구함
        return node.data if node else None    # 노드가 있는/없는 경우 각각 데이터/None 반환

    def insert(self, pos, elem):
        if pos < 0 or pos > self.length:      # 잘못된 위치인 경우
            return None                       # None 반환
        
        node = Node(elem)                     # 삽입할 새로운 노드를 만듦
        if pos == 0:                          # 머리노드로 삽입하는 경우
            node.next = self.head             # node의 링크가 머리노드를 가리킴
            self.head = node                  # 이제 node가 머리노드가 됨
        else:                                 # 머리노드가 아닌 경우
            before = self.getNode(pos-1)      # 삽입할 위치 이전 노드 탐색
            before.append(node)               # before 다음에 추가

        self.length += 1                      # 요소 추가시 길이 증가

    def delete(self, pos):
        if self.isEmpty() or pos < 0 or pos >= self.length:
            return None

        if pos == 0:                          # 머리노드 삭제 경우
            removed_node = self.head  
            self.head = self.head.next        # 머리노드를 갱신
        else:
            before = self.getNode(pos-1)      # 삭제할 위치 이전 노드 탐색
            removed_node = before.popNext()   # before의 다음 노드 삭제
            
        self.length -= 1                      # 요소 삭제시 길이 감소
        return removed_node.data if removed_node else None
    
    # 리스트 상태를 한눈에 보기 위한 보조 메서드
    def display(self, msg='LinkedList:'):
        print(msg, end=' ')
        ptr = self.head
        while ptr is not None:
            print(ptr.data, end=' -> ')
            ptr = ptr.next
        print('None')
```

```python
L = LinkedList()
L.insert(0, "Apple")    # Apple -> None
L.insert(1, "Banana")   # Apple -> Banana -> None
L.insert(1, "Cherry")   # Apple -> Cherry -> Banana -> None
L.display("삽입 후:")     # 결과: Apple -> Cherry -> Banana -> None

print(f"현재 노드 개수: {len(L)}") # 결과: 3

L.delete(0)             # 머리(Apple) 삭제
L.display("삭제 후:")     # 결과: Cherry -> Banana -> None
print(f"현재 노드 개수: {len(L)}") # 결과: 2
```

## 2-2. 이중 연결 리스트 구현

```python
class DNode:                                          # 이중 연결 노드 클래스
    def __init__ (self, elem, prev=None, next=None):
        self.data = elem                              # 노드의 데이터 필드(요소)
        self.next = next                              # 다음노드를 위한 링크
        self.prev = prev                              # 이전노드를 위한 링크(추가됨)

    def append(self, node):                           # self 다음에 node를 넣는 연산
        if node is not None:                          # node가 None이 아니면
            node.next = self.next                     # ① 수행
            node.prev = self                          # ② 수행
            if node.next is not None:                 # ③ self의 다음 노드가 있으면
                node.next.prev = node                 #    그 노드의 이전 노드를 node로 수정
            self.next = node                          # ④ 수행

    def popNext(self):                                # self 다음노드 삭제 연산
        node = self.next                              # 삭제할 노드
        if node is not None:                          # next가 None이 아니면
            self.next = node.next                     # ① 수행
            if self.next is not None:                 # ② 다음 노드가 있으면
                self.next.prev = self                 # 그 노드의 이전노드는 self
        return node                                   # 삭제할 노드 반환
```

```python
class DblLinkedList:                                  # 이중연결리스트 클래스
    def __init__(self):                               # 생성자
        self.head = None                              # head 선언 및 None으로 초기화
        self.length = 0                               # length 선언 및 0으로 초기화

    def __len__(self):
        return self.length                            # len() 함수 사용시 length 반환

    def isEmpty(self):                  
        return self.head is None       

    def isFull(self):                
        return False                    

    def getNode(self, pos):
        if pos < 0 or pos >= self.length:     
            return None                   
        ptr = self.head                     
        for _ in range(pos):                
            ptr = ptr.next                   
        return ptr                      

    def getEntry(self, pos):
        node = self.getNode(pos)           
        return node.data if node else None    

    def insert(self, pos, elem):
        if pos < 0 or pos > self.length:               # 잘못된 위치인 경우
            return None                                # None 반환

        node = DNode(elem)                             # 삽입할 새로운 노드를 만듦
        if pos == 0:                                   # 머리노드로 삽입하는 경우
            node.next = self.head                      # node의 링크가 머리노드를 가리킴
            if node.next is not None:                  # node 다음 노드가 있으면
                node.next.prev = node                  # 그 노드의 이전노드는 node
            self.head = node                           # 이제 node가 머리노드가 됨
        else:                                          # 머리노드가 아닌 경우
            before = self.getNode(pos-1)               # 삽입할 위치 이전 노드 탐색
            before.append(node)                        # before 다음에 추가

        self.length += 1                               # 요소 추가시 길이 증가

    def delete(self, pos):
        if self.isEmpty() or pos < 0 or pos >= self.length:
            return None
        
        if pos == 0:                                   # 머리노드 삭제 경우
            removed_node = self.head  
            self.head = self.head.next                 # 머리노드를 갱신
            if self.head is not None:
                self.head.prev = None                  # 머리노드는 이전노드 없음
        else:
            before = self.getNode(pos-1)               # 삭제할 위치 이전 노드 탐색
            removed_node = before.popNext()            # before의 다음 노드 삭제
            
        self.length -= 1                               # 요소 삭제시 길이 감소
        return removed_node.data if removed_node else None
    
    # 리스트 상태를 한눈에 보기 위한 보조 메서드
    def display_forward(self, msg='DblLinkedList Forward:'):
        print(msg, end=' ')
        ptr = self.head
        while ptr is not None:
            print(ptr.data, end=' -> ')
            ptr = ptr.next
        print('None')

    def display_backward(self, msg='DblLinkedList Backward:'):
        print(msg, end=' ')
        ptr = self.head
        while ptr.next is not None:                     # 꼬리노드로 이동
            ptr = ptr.next

        while ptr is not None:
            print(ptr.data, end=' <- ')
            ptr = ptr.prev
        print('None')
```

```python
DL = DblLinkedList()
DL.insert(0, "Apple")    # Apple <-> None
DL.insert(1, "Banana")   # Apple <-> Banana <-> None
DL.insert(1, "Cherry")   # Apple <-> Cherry <-> Banana <-> None

DL.display_forward("삽입 후(정방향):")  # 결과: Apple -> Cherry -> Banana -> None
DL.display_backward("삽입 후(역방향):") # 결과: Banana <- Cherry <- Apple <- None

print(f"현재 노드 개수: {len(DL)}") # 결과: 3

DL.delete(0)
DL.display_forward("삭제 후(정방향):")   # 결과: Cherry -> Banana -> None
DL.display_backward("삭제 후(역방향):")  # 결과: Banana <- Cherry <- None
print(f"현재 노드 개수: {len(DL)}") # 결과: 2
```

## 2-3. 연습 문제
[문제명] 표 편집
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/81303 

```python
class DNode:
    def __init__(self, elem, prev=None, next=None):
        self.data = elem
        self.next = next
        self.prev = prev

    def append(self, node):
        if node is not None:
            node.next = self.next
            node.prev = self
            if node.next is not None:
                node.next.prev = node
            self.next = node

    def popNext(self):
        node = self.next
        if node is not None:
            self.next = node.next
            if self.next is not None:
                self.next.prev = self
        return node

class DblLinkedList:
    def __init__(self, n, k):
        # 노드 객체 생성
        self.nodes = [DNode(i) for i in range(n)]
        
        # 연결 설정
        for i in range(n):
            if i > 0:
                self.nodes[i].prev = self.nodes[i-1]
            if i < n - 1:
                self.nodes[i].next = self.nodes[i+1]
        
        self.curr = self.nodes[k]
        self.deleted = []

    # U 연산    
    def up(self, x):
        for _ in range(x):
            if self.curr.prev: 
                self.curr = self.curr.prev

    # D 연산
    def down(self, x):
        for _ in range(x):
            if self.curr.next:
                self.curr = self.curr.next

    # C 연산
    def remove(self):
        target = self.curr
        self.deleted.append(target)
        
        prev_node = target.prev
        next_node = target.next
        
        # popNext를 활용하여 target 노드 삭제
        if prev_node:
            prev_node.popNext()
        else:
            # 첫 번째 노드인 경우 (prev가 없음) 수동으로 다음 노드의 prev 연결 끊기
            if next_node:
                next_node.prev = None
        
        # 삭제 후 선택 노드 이동 (마지막 노드면 위쪽, 아니면 아래쪽)
        self.curr = next_node if next_node else prev_node

    # Z 연산
    def recover(self):
        node = self.deleted.pop()
        prev_node = node.prev
        
        # append를 활용하여 복구
        if prev_node:
            prev_node.append(node)
        else:
            # 첫 번째 노드로 복구되는 경우 (prev가 없음) 원래 next였던 노드의 prev를 자신으로 연결
            if node.next:
                node.next.prev = node

def solution(n, k, cmd):
    table = DblLinkedList(n, k)
    
    for c in cmd:
        command = c.split()
        if command[0] == 'U':
            table.up(int(command[1]))
        elif command[0] == 'D':
            table.down(int(command[1]))
        elif command[0] == 'C':
            table.remove()
        elif command[0] == 'Z':
            table.recover()
            
    # 삭제된 노드 번호
    deleted_nums = {node.data for node in table.deleted}
    
    result = []
    for i in range(n):
        result.append('X' if i in deleted_nums else 'O')
        
    return "".join(result)
```

```python
print(solution(n=8, k=2, cmd=["D 2","C","U 3","C","D 4","C","U 2","Z","Z"]))  # 실행결과: OOOOXOOO
print(solution(n=8, k=2, cmd=["D 2","C","U 3","C","D 4","C","U 2","Z","Z","U 1","C"]))  # 실행결과: OOXOXOOO
```
