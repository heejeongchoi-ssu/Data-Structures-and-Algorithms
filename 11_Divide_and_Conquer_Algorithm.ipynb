# **1. 병합 정렬**
## 1-1. 병합 정렬 구현

```python
def merge_sort(A, left, right, sorted):       # A[left...right]를 오름차순으로 정렬
    if left < right:                          # 항목이 2개 이상인 경우
        mid = (left + right) // 2             # 리스트의 균등 분할
        merge_sort(A, left, mid, sorted)      # 왼쪽 부분 리스트 정렬
        merge_sort(A, mid + 1, right, sorted) # 오른쪽 부분 리스트 정렬
        merge(A, left, mid, right, sorted)    # 병합
    # else: 항목이 1개인 경우 자동으로 정복됨

def merge(A, left, mid, right, sorted):
    k = left                                  # 병합을 위한 임시 리스트의 인덱스
    i = left                                  # 왼쪽 리스트의 인덱스
    j = mid + 1                               # 오른쪽 리스트의 인덱스
    
    while i <= mid and j <= right:
        if A[i] <= A[j]:	
            sorted[k] = A[i]
            i, k = i + 1, k + 1
        else:
            sorted[k] = A[j]
            j, k = j + 1, k + 1

    # 한쪽에 남아 있는 레코드의 일괄 복사
    if i > mid:
        sorted[k:right + 1] = A[j:right + 1]   # 오른쪽 리스트의 남은 원소 복사
    else:
        sorted[k:right + 1] = A[i:mid + 1]     # 왼쪽 리스트의 남은 원소 복사

    A[left:right + 1] = sorted[left:right + 1] # A로 복사
```

```python
A = [5, 3, 8, 4, 9, 1, 6, 2, 7]
merge_sort(A, left=0, right=len(A)-1, sorted=[0]*len(A))
print(A)
```

## 1-2. 연습 문제
[문제명] K번째 수
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/42748

```python
def solution(array, commands):
    answer = []
    for cmd in commands:
        i, j, k = cmd
        # 1. 자르기 (인덱스는 0부터 시작하므로 i-1부터 j까지)
        sub_array = array[i-1:j]
        
        # 2. 병합 정렬(Merge Sort) 수행
        n = len(sub_array)
        merge_sort(sub_array, left=0, right=n-1, sorted=[0]*n)
        
        # 3. k번째 수 추출 (인덱스는 k-1)
        answer.append(sub_array[k-1])
    return answer
```

```python
print(solution(array=[1, 5, 2, 6, 3, 7, 4], commands=[[2, 5, 3], [4, 4, 1], [1, 7, 3]]))
```

# **2. 퀵 정렬**
## 2-1. 퀵 정렬 구현

```python
def quick_sort(A, left, right):                 # A[left...right]를 오름차순으로 정렬
    if left < right:                            # 항목이 2개 이상인 경우
        pivot_idx = partition(A, left, right)   # A의 피벗 인덱스
        quick_sort(A, left, pivot_idx - 1)      # 왼쪽 부분 리스트 정렬
        quick_sort(A, pivot_idx + 1, right)     # 오른쪽 부분 리스트 정렬

def partition(A, left, right):
	pivot = A[left]                             # 피벗 설정
	low = left + 1                              # 왼쪽 부분 리스트의 인덱스(증가 방향)
	high = right                                # 오른쪽 부분 리스트의 인덱스(감소 방향)

	while low < high:                           # low와 high가 역전되지 않는 한 반복
		while low <= right and A[low] <= pivot:
			low += 1                            # low 증가
		while high >= left and A[high] > pivot:
			high -= 1                           # high 감소
		
		if low < high:			                # 조건에 맞지 않는 요소 교환
			A[low], A[high] = A[high], A[low]
	
	A[left], A[high] = A[high], A[left]	        # 마지막으로 high와 피벗 항목 교환 
	return high					                # 피벗 인덱스 반환
```

```python
A = [5, 3, 8, 4, 9, 1, 6, 2, 7]
quick_sort(A, left=0, right=len(A)-1)
print(A)
```

## 2-2. 연습 문제
[문제명] K번째 수
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/42748

```python
def solution(array, commands):
    answer = []
    for cmd in commands:
        i, j, k = cmd
        # 1. 자르기 (인덱스는 0부터 시작하므로 i-1부터 j까지)
        sub_array = array[i-1:j]
        
        # 2. 퀵 정렬(Quick Sort) 수행
        n = len(sub_array)
        quick_sort(sub_array, left=0, right=n-1)
        
        # 3. k번째 수 추출 (인덱스는 k-1)
        answer.append(sub_array[k-1])
    return answer
```

```python
print(solution(array=[1, 5, 2, 6, 3, 7, 4], commands=[[2, 5, 3], [4, 4, 1], [1, 7, 3]]))
```
