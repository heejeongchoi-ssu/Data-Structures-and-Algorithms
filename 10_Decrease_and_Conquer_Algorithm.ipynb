# **1. 삽입 정렬**
## 1-1. 삽입 정렬 구현

```python
def insertion_sort(A) :
    n = len(A)
    for i in range(1, n) :             # i범위: 1~n-1
        key = A[i]                     # A[i]를 key에 저장
        j = i - 1                      # 탐색은 i-1부터 앞으로 진행
        while j >= 0 and A[j] > key :  # key보다 작은 A[j]가 나올 때 까지 
            A[j + 1] = A[j]            # A[j]를 미리 뒤로 한 칸 옮김
            j -= 1
        A[j + 1] = key                 # j+1이 A[i]가 삽입될 위치임
```

```python
print(insertion_sort(A=[5, 3, 8, 4, 9, 1, 6, 2, 7]))
```

## 1-2. 연습 문제
[문제명] K번째 수
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/42748

```python
def solution(array, commands):
    answer = []
    for cmd in commands:
        i, j, k = cmd
        # 1. 자르기 (인덱스는 0부터 시작하므로 i-1부터 j까지)
        sub_array = array[i-1:j]
        
        # 2. 삽입 정렬(Insertion Sort) 수행
        n = len(sub_array)
        for i in range(1, n):
            key = sub_array[i]
            j = i - 1
            while j >= 0 and sub_array[j] > key:
                sub_array[j + 1] = sub_array[j]
                j -= 1
            sub_array[j + 1] = key
        
        # 3. k번째 수 추출 (인덱스는 k-1)
        answer.append(sub_array[k-1])
    return answer
```

```python
print(solution(array=[1, 5, 2, 6, 3, 7, 4], commands=[[2, 5, 3], [4, 4, 1], [1, 7, 3]]))
```

# **2. 이진 탐색**
## 2-1. 이진 탐색 구현 (재귀)

```python
def binary_search_recur(A, key, low, high):
    if low <= high:                                           # 항목들이 남아 있으면(종료 조건)
        mid = (low + high) // 2                               # 중앙 인덱스 계산
        if key == A[mid]:                                     # 탐색 성공하면
            return mid                                        # 인덱스 반환
        elif key < A[mid]:                                    # 왼쪽 부분 리스트 탐색
            return binary_search_recur(A, key, low, mid - 1)
        else:                                                 # 오른쪽 부분 리스트 탐색
            return binary_search_recur(A, key, mid + 1, high)
    return -1                                                 # 탐색 실패시 -1 반환
```

```python
print(binary_search_recur(A=[5, 3, 8, 4, 2, 7], key=4, low=0, high=6))
print(binary_search_recur(A=[5, 3, 8, 4, 2, 7], key=6, low=0, high=6))
```

## 2-2. 이진 탐색 구현 (반복)

```python
def binary_search_iter(A, key, low, high):
    while low <= high:                 # 항목들이 남아 있으면(종료 조건)
        mid = (low + high) // 2        # 중앙 인덱스 계산
        if key == A[mid]:              # 탐색 성공하면
            return mid                 # 인덱스 반환
        elif key < A[mid]:             # 왼쪽 부분 리스트 탐색
            high = mid - 1             # low ~ mid - 1 범위 탐색
        else:                          # 오른쪽 부분 리스트 탐색
            low = mid + 1              # mid+1 ~ high 범위 탐색
    return -1                          # 탐색 실패시 -1 반환
```

```python
print(binary_search_iter(A=[5, 3, 8, 4, 2, 7], key=4, low=0, high=6))
print(binary_search_iter(A=[5, 3, 8, 4, 2, 7], key=6, low=0, high=6))
```

## 2-3. 연습 문제
[문제명] 입국심사
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/43238 

```python
def solution(n, times):
    left = 1  # 최소 심사 시간
    right = min(times) * n  # 탐색할 최대 시간
    
    while left <= right:  # 이진 탐색 수행
        mid = (left + right) // 2

        # mid 시간 동안 모든 심사관이 처리할 수 있는 총 사람 수
        people = sum(mid // time for time in times)
        
        if people >= n:  # mid 시간 안에 n명 이상 심사할 수 있다면
            right = mid - 1  # 더 짧은 시간에도 가능한지 왼쪽 범위를 탐색
        else:  # mid 시간 안에 n명을 모두 심사할 수 없다면
            left = mid + 1  # 시간이 부족하므로 더 긴 시간을 탐색
    return left
```

```python
print(solution(n=6, times=[7, 10]))
```
