# **1. 스택**
## 1-1. 스택 구현

```python
class ArrayStack:
    def __init__(self, capacity):	            # 생성자 정의
        self.capacity = capacity		          # 용량(고정)
        self.array = [None] * self.capacity   # 요소 배열: [None, …, None]
        self.top = -1			                    # 스택 상단의 인덱스: 공백 상태(-1)

    def is_empty(self):
        return self.top == -1

    def is_full(self):
        return self.top == self.capacity - 1

    def push(self, item):
        if self.is_full():                    # 포화 상태인 경우(overflow)
            print("Stack is Full!")
            return
        self.top += 1                         # top 증가
        self.array[self.top] = item           # top 위치에 item 삽입

    def pop(self):
        if self.is_empty():                   # 공백 상태인 경우(underflow)
            print("Stack is Empty!")
            return None
        self.top -= 1                         # top 감소
        item = self.array[self.top + 1]       # 제거할 요소
        self.array[self.top + 1] = None       # 배열에서 요소 제거
        return item

    def peek(self):
        if self.is_empty():
            print("Stack is Empty!")
            return None
        return self.array[self.top]

    def size(self):
        return self.top + 1                   # 현재 요소의 수는 top + 1
```

```python
class ListStack:
    def __init__(self):
        self.stack = []  # 빈 리스트로 초기화

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)  # 리스트 맨 뒤에 추가

    def pop(self):
        if self.is_empty():
            print("Stack is Empty!")
            return None
        return self.stack.pop()  # 리스트의 마지막 요소 제거 및 반환

    def peek(self):
        if self.is_empty():
            print("Stack is Empty!")
            return None
        return self.stack[-1]  # 리스트의 마지막 요소 반환

    def size(self):
        return len(self.stack)
```

## 1-2. 연습 문제
[문제명] 올바른 괄호
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/12909 

```python
def solution(s):
    stack = []
    for i in s:
        if i == '(':
            stack.append(i)
        else:
            if len(stack) == 0:
                return False
            else:
                stack.pop()
                
    return len(stack) == 0
```

# **2. 큐**
## 2-1. 선형 큐 구현

```python
class ArrayQueue:
    def __init__(self, capacity):                # 생성자 정의
        self.capacity = capacity                 # 용량(고정)
        self.array = [None] * capacity           # 요소들을 저장할 배열
        self.front = -1                          # 전단 인덱스
        self.rear = -1                           # 후단 인덱스

    def is_empty(self):
        return self.front == self.rear

    def is_full(self):
        return self.rear == self.capacity - 1

    def enqueue(self, item):
        if self.is_full():                       # 포화 상태인 경우(overflow)
            print("Queue is Full!")
            return
        self.rear += 1                           # rear 증가
        self.array[self.rear] = item             # rear 위치에 item 삽입

    def dequeue(self):
        if self.is_empty():                      # 공백 상태인 경우(underflow)
            print("Queue is Empty!")
            return None
        self.front += 1                          # front 증가
        item = self.array[self.front]            # 제거할 요소
        self.array[self.front] = None            # 배열에서 요소 제거
        return item

    def peek(self):
        if self.is_empty():                      # 공백 상태인 경우(underflow)
            print("Queue is Empty!")
            return None
        return self.array[self.front + 1]        # front + 1 위치의 요소 반환

    def size(self):
        return self.rear - self.front            # 현재 요소의 수는 rear - front
```

```python
class ListQueue:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if self.is_empty():
            print("Queue is Empty!")
            return None
        return self.queue.pop(0)

    def peek(self):
        if self.is_empty():
            print("Queue is Empty!")
            return None
        return self.queue[0]

    def size(self):
        return len(self.queue)
```

```python
from collections import deque

class DequeQueue:
    def __init__(self):
        self.queue = deque()

    def is_empty(self):
        return len(self.queue) == 0

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if self.is_empty():
            print("Queue is Empty!")
            return None
        return self.queue.popleft()

    def peek(self):
        if self.is_empty():
            print("Queue is Empty!")
            return None
        return self.queue[0]

    def size(self):
        return len(self.queue)
```

## 2-2. 원형 큐 구현

```python
class CircularQueue :
    def __init__(self, capacity):                            # 생성자 정의
        self.capacity = capacity                             # 용량(고정)
        self.array = [None] * capacity                       # 요소들을 저장할 배열
        self.front = 0                                       # 첫 번째 요소의 바로 앞 인덱스
        self.rear = 0                                        # 마지막 요소의 인덱스

    def is_empty(self):
        return self.front == self.rear

    def is_full(self):
        return self.front == (self.rear + 1) % self.capacity

    def enqueue(self, item):
        if self.is_full():                                   # 포화 상태인 경우(overflow)
            print("Queue is Full!")
            return
        self.rear = (self.rear + 1) % self.capacity          # rear 증가 및 시계방향회전 반영
        self.array[self.rear] = item                         # rear 위치에 item 삽입

    def dequeue(self):
        if self.is_empty():                                  # 공백 상태인 경우(underflow)
            print("Queue is Empty!")
            return None
        self.front = (self.front + 1) % self.capacity        # front 증가 및 시계방향회전 반영
        item = self.array[self.front]                        # 제거할 요소
        self.array[self.front] = None                        # 배열에서 요소 제거
        return item

    def peek(self):
        if self.is_empty():                                  # 공백 상태인 경우(underflow)
            print("Queue is Empty!")
            return None
        return self.array[(self.front + 1) % self.capacity]  # front + 1 위치의 요소 반환

    def size(self):
        return (self.rear - self.front + self.capacity) % self.capacity
```

## 2-3. 데크 구현

```python
class ArrayDeque:
    def __init__(self, capacity):
        self.capacity = capacity
        self.array = [None] * capacity
        self.front = -1
        self.rear = -1

    def is_empty(self):
        return self.front == self.rear

    def is_full(self):
        return self.rear == self.capacity - 1

    def size(self):  # ArrayQueue의 size와 동일
        return self.rear - self.front

    def delete_front(self):  # ArrayQueue의 dequeue와 동일
        if self.isEmpty():
            print("Deque is Empty!")
            return None
        self.front += 1
        item = self.array[self.front]
        self.array[self.front] = None
        return item

    def get_front(self):  # ArrayQueue의 peek과 동일
        if self.isEmpty():
            print("Deque is Empty!")
            return None
        return self.array[self.front + 1]

    def add_rear(self, item):  # ArrayQueue의 enqueue와 동일
        if self.isFull():
            print("Deque is Full!")
            return
        self.rear += 1
        self.array[self.rear] = item

    def delete_rear(self):  # ArrayStack의 pop과 동일
        if self.isEmpty():
            print("Deque is Empty!")
            return None
        self.rear -= 1
        item = self.array[self.rear + 1]
        self.array[self.rear + 1] = None
        return item

    def get_rear(self):  # ArrayStack의 peek과 동일
        if self.isEmpty():
            print("Deque is Empty!")
            return None
        return self.array[self.rear]

    def add_front(self, item):
        if self.isFull():
            print("Deque is Full!")
            return

        for i in range(self.rear, self.front, -1):  # 모든 요소를 한 칸씩 뒤로 이동
            self.array[i + 1] = self.array[i]

        self.array[self.front + 1] = item           # front + 1 위치에 요소 삽입
        self.rear += 1                              # rear 증가
```

```python
from collections import deque

class Deque:
    def __init__(self):
        self.deque = deque()

    def is_empty(self):
        return len(self.deque) == 0

    def size(self):
        return len(self.deque)

    def add_front(self, item):
        self.deque.appendleft(item)

    def add_rear(self, item):
        # 기존 코드의 rear 삽입 로직
        self.deque.append(item)

    def delete_front(self):
        if self.is_empty():
            print("Deque is Empty!")
            return None
        return self.deque.popleft()

    def delete_rear(self):
        if self.is_empty():
            print("Deque is Empty!")
            return None
        return self.deque.pop()

    def get_front(self):
        if self.is_empty():
            print("Deque is Empty!")
            return None
        return self.deque[0]

    def get_rear(self):
        if self.is_empty():
            print("Deque is Empty!")
            return None
        return self.deque[-1]
```

## 2-4. 연습 문제
[문제명] 다리를 지나는 트럭
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/42583

```python
from collections import deque

def solution(bridge_length, weight, truck_weights):
    time = 0
    bridge = deque([0] * bridge_length)
    truck_weights = deque(truck_weights)  # 트럭 대기 큐
    current_weight = 0

    while bridge:
        time += 1
        out_truck = bridge.popleft()
        current_weight -= out_truck
        
        if truck_weights:  
            if current_weight + truck_weights[0] <= weight:  # 새로운 트럭이 올라갈 수 있는지 확인
                in_truck = truck_weights.popleft()
                bridge.append(in_truck)
                current_weight += in_truck
            else:  # 못 올라가면 0 추가 (다리 길이 유지)
                bridge.append(0)
    return time
```
