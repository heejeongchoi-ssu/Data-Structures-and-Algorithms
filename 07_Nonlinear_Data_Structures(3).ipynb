# **1. 깊이 우선 탐색(DFS)**
## 1-1. DFS 구현 - 반복

```python
def dfs_iterative(graph, start_v):
    visited = []  # 시작 정점을 방문 처리
    stack = [start_v]

    while stack:
        v = stack.pop()
        if v not in visited:
            visited.append(v)  # 방문 처리
            for adj_v in reversed(graph[v]):  # 인접한 정점 방문
                if adj_v not in visited:
                    stack.append(adj_v)
    return visited
```

```python
# 그래프 정의 (인접 리스트 방식)
graph = {
    'U': ['V', 'W'],
    'V': ['U', 'W', 'X'],
    'W': ['U', 'V', 'Y'],
    'Y': ['W'],
    'X': ['V']
}
```

```python
dfs_iterative(graph, 'U')
```

## 1-2. DFS 구현 - 재귀

```python
def dfs_recursive(graph, v, visited):
    visited.append(v)

    for adj_v in graph[v]:
        if adj_v not in visited:
            dfs_recursive(graph, adj_v, visited)
    return visited
```

```python
dfs_recursive(graph, 'U', [])
```

## 1-3. 연습 문제
[문제명] 네트워크
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/43162

```python
def solution(n, computers):
    answer = 0
    visited = [False] * n  # 방문 여부 리스트

    def dfs(v):
        visited[v] = True  # 현재 컴퓨터 방문 처리
        for neighbor in range(n):
            # 연결되어 있고 아직 방문하지 않은 컴퓨터라면 탐색 진행
            if computers[v][neighbor] == 1 and not visited[neighbor]:
                dfs(neighbor)

    for i in range(n):
        if not visited[i]:  # 아직 어떤 네트워크에도 속하지 않은 컴퓨터라면
            dfs(i)          # 새로운 네트워크 탐색 시작
            answer += 1     # 탐색이 끝나면 네트워크 개수 추가
            
    return answer
```

```python
print(solution(n=3, computers=[[1, 1, 0], [1, 1, 0], [0, 0, 1]]))
print(solution(n=3, computers=[[1, 1, 0], [1, 1, 1], [0, 1, 1]]))
```

# **2. 너비 우선 탐색(BFS)**
## 2-1. BFS 구현 - 반복

```python
from collections import deque

def bfs_iterative(graph, start_v):
    visited = []
    queue = deque([start_v])

    while queue:
        v = queue.popleft()
        if v not in visited:
            visited.append(v)  # 방문 처리

            for adj_v in reversed(graph[v]):  # 인접한 정점 방문
                if adj_v not in visited:
                    queue.append(adj_v)
    return visited
```

```python
bfs_iterative(graph, 'U')
```

## 2-2. 연습 문제
[문제명] 단어 변환
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/43163

```python
from collections import deque

def can_convert(word1, word2):  # 두 단어가 한 글자만 다른지 확인하는 함수
    diff_count = 0
    for c1, c2 in zip(word1, word2):
        if c1 != c2:
            diff_count += 1
    return diff_count == 1

def solution(begin, target, words):
    # 목표 단어가 단어 리스트에 없으면 도달 불가능
    if target not in words:
        return 0

    # (현재 단어, 변환 횟수)를 큐에 저장
    queue = deque([(begin, 0)])
    visited = set([begin])

    while queue:
        current_word, step = queue.popleft()

        # 목표 단어에 도달하면 현재까지의 단계 반환
        if current_word == target:
            return step

        # 전체 단어 리스트를 순회하며 변환 가능한 단어 찾기
        for word in words:
            if word not in visited and can_convert(current_word, word):
                visited.add(word)
                queue.append((word, step + 1))

    return 0  # 변환할 수 없는 경우
```

```python
print(solution(begin="hit", target="cog", words=["hot", "dot", "dog", "lot", "log", "cog"]))
print(solution(begin="hit", target="cog", words=["hot", "dot", "dog", "lot", "log"]))
```
