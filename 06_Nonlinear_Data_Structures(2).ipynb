# **1. 힙**
## 1-1. 최소 힙 구현

```python
class MinHeap:
    def __init__(self, capacity):
        self.capacity = capacity
        self.arr = [None] * (capacity + 1)  # 인덱스 1부터 사용하기 위해 capacity + 1 크기로 할당
        self._size = 0

    def is_empty(self):
        return self._size == 0

    def is_full(self):
        return self._size == self.capacity

    def size(self):
        return self._size

    def peek(self):
        if self.is_empty():
            print("Heap is empty!")
            return None
        return self.arr[1]

    def insert(self, key):
        if self.is_full():
            print("Heap is full!")
            return False

        self._size += 1
        self.arr[self._size] = key  # 새 데이터를 마지막 위치에 추가
        self._sift_up(self._size)  # 새로운 노드를 부모와 비교하며 적절한 위치까지 위로 이동
        return True

    def _sift_up(self, idx):
        while idx > 1:  # 새로운 노드를 부모와 비교하며 적절한 위치까지 위로 이동
            parent = idx // 2  # 부모 노드의 인덱스 계산
            if self.arr[idx] < self.arr[parent]:  # 최소 힙 규칙 위반
                self.arr[idx], self.arr[parent] = self.arr[parent], self.arr[idx]  # swap
                idx = parent  # 인덱스를 부모 위치로 갱신하여 계속 위로 이동
            else:  # 최소 힙 규칙 만족시 종료
                break

    def delete(self):
        if self.is_empty():
            print("Heap is empty!")
            return None

        root = self.arr[1]  # 루트 노드 저장
        self.arr[1] = self.arr[self._size]  # 마지막 노드를 루트 노드로 이동
        self.arr[self._size] = None  # 마지막 노드 제거
        self._size -= 1

        if not self.is_empty():
            self._sift_down(1)  # 자식들과 비교하며 적절한 위치까지 아래로 이동
        return root

    def _sift_down(self, idx):
        while idx * 2 <= self._size:  # 자식들과 비교하며 적절한 위치까지 아래로 이동
            left = idx * 2
            right = idx * 2 + 1
            smallest = idx  # 현재 노드, 왼쪽 자식, 오른쪽 자식 중 가장 작은 값

            if self.arr[left] < self.arr[smallest]:  # 왼쪽 자식과 현재 노드 비교
                smallest = left

            if right <= self._size and self.arr[right] < self.arr[smallest]:  # 오른쪽 자식과 현재 노드 비교
                smallest = right

            if smallest != idx:  # 최소 힙 규칙 위반 (현재 노드가 최소 값 아님)
                self.arr[idx], self.arr[smallest] = self.arr[smallest], self.arr[idx]
                idx = smallest  # 인덱스를 자식 위치로 갱신하여 계속 아래로 이동
            else:  # 최소 힙 규칙 만족시 종료
                break

    def heapify(self, input_list):
        if len(input_list) > self.capacity:
            print("Input list exceeds capacity!")
            return False
        
        for i in range(len(input_list)):  # 입력받은 데이터를 배열로 복사
            self.arr[i + 1] = input_list[i]
        
        self._size = len(input_list)

        # 마지막 부모 노드(size // 2)부터 루트(1)까지 역순으로 _sift_down 수행
        # 리프 노드는 자식 노드가 없어서 이미 힙 규칙을 만족하므로 건너뜀
        for i in range(self._size // 2, 0, -1):
            self._sift_down(i)
        return True
```

```python
heap = MinHeap(10)

heap.insert(50)
heap.insert(20)
heap.insert(70)
print(heap.arr)

heap.delete()
print(heap.arr)
```

```python
heap = MinHeap(10)

heap.heapify([50, 20, 70])
print(heap.arr)
```

## 1-2. 최소 힙 구현 - heapq 모듈 활용

```python
import heapq

class MinHeap:
    def __init__(self):
        self.heap = []

    def is_empty(self):
        return len(self.heap) == 0

    def size(self):
        return len(self.heap)

    def peek(self):
        if self.is_empty():
            print("Heap is empty!")
            return None
        return self.heap[0]

    def insert(self, key):
        heapq.heappush(self.heap, key)

    def delete(self):
        if self.is_empty():
            print("Heap is empty!")
            return None

        root = heapq.heappop(self.heap)
        return root

    def heapify(self, input_list):
        self.heap = input_list
        heapq.heapify(self.heap)
```

```python
heap = MinHeap()

heap.insert(50)
heap.insert(20)
heap.insert(70)
print(heap.heap)

heap.delete()
print(heap.heap)
```

```python
heap = MinHeap()

heap.heapify([50, 20, 70])
print(heap.heap)
```

```python
class MaxHeap:
    def __init__(self, capacity):
        self.capacity = capacity
        self.arr = [None] * (capacity + 1)  # 인덱스 1부터 사용하기 위해 capacity + 1 크기로 할당
        self._size = 0

    def is_empty(self):
        return self._size == 0

    def is_full(self):
        return self._size == self.capacity

    def size(self):
        return self._size

    def peek(self):
        if self.is_empty():
            print("Heap is empty!")
            return None
        return self.arr[1]

    def insert(self, key):
        if self.is_full():
            print("Heap is full!")
            return False

        self._size += 1
        self.arr[self._size] = key  # 새 데이터를 마지막 위치에 추가
        self._sift_up(self._size)  # 새로운 노드를 부모와 비교하며 적절한 위치까지 위로 이동
        return True

    def _sift_up(self, idx):
        while idx > 1:  # 새로운 노드를 부모와 비교하며 적절한 위치까지 위로 이동
            parent = idx // 2  # 부모 노드의 인덱스 계산
            if self.arr[idx] > self.arr[parent]:  # 최대 힙 규칙 위반
                self.arr[idx], self.arr[parent] = self.arr[parent], self.arr[idx]  # swap
                idx = parent  # 인덱스를 부모 위치로 갱신하여 계속 위로 이동
            else:  # 최대 힙 규칙 만족시 종료
                break

    def delete(self):
        if self.is_empty():
            print("Heap is empty!")
            return None

        root = self.arr[1]  # 루트 노드 저장
        self.arr[1] = self.arr[self._size]  # 마지막 노드를 루트 노드로 이동
        self.arr[self._size] = None  # 마지막 노드 제거
        self._size -= 1

        if not self.is_empty():
            self._sift_down(1)  # 자식들과 비교하며 적절한 위치까지 아래로 이동
        return root

    def _sift_down(self, idx):
        while idx * 2 <= self._size:  # 자식들과 비교하며 적절한 위치까지 아래로 이동
            left = idx * 2
            right = idx * 2 + 1
            largest = idx  # 현재 노드, 왼쪽 자식, 오른쪽 자식 중 가장 큰 값

            if self.arr[left] > self.arr[largest]:  # 왼쪽 자식과 현재 노드 비교
                largest = left

            if right <= self._size and self.arr[right] > self.arr[largest]:  # 오른쪽 자식과 현재 노드 비교
                largest = right

            if largest != idx:  # 최대 힙 규칙 위반 (현재 노드가 최대 값 아님)
                self.arr[idx], self.arr[largest] = self.arr[largest], self.arr[idx]
                idx = largest  # 인덱스를 자식 위치로 갱신하여 계속 아래로 이동
            else:  # 최대 힙 규칙 만족시 종료
                break

    def heapify(self, input_list):
        if len(input_list) > self.capacity:
            print("Input list exceeds capacity!")
            return False
        
        for i in range(len(input_list)):  # 입력받은 데이터를 배열로 복사
            self.arr[i + 1] = input_list[i]
        
        self._size = len(input_list)

        # 마지막 부모 노드(size // 2)부터 루트(1)까지 역순으로 _sift_down 수행
        # 리프 노드는 자식 노드가 없어서 이미 힙 규칙을 만족하므로 건너뜀
        for i in range(self._size // 2, 0, -1):
            self._sift_down(i)
        return True
```

## 1-4. 최대 힙 구현 - heapq 모듈 활용

```python
import heapq

class MaxHeap:
    def __init__(self):
        self.heap = []

    def is_empty(self):
        return len(self.heap) == 0

    def size(self):
        return len(self.heap)

    def peek(self):
        if self.is_empty():
            print("Heap is empty!")
            return None
        return -1 * self.heap[0]

    def insert(self, key):
        heapq.heappush(self.heap, -1 * key)  # 삽입 시 음수로 변환

    def delete(self):
        if self.is_empty():
            print("Heap is empty!")
            return None

        root = heapq.heappop(self.heap)
        return -1 * root

    def heapify(self, input_list):
        self.heap = [-1 * i for i in input_list]
        heapq.heapify(self.heap)
```

```python
heap = MaxHeap()

heap.insert(50)
heap.insert(20)
heap.insert(70)
print(heap.heap)

heap.delete()
print(heap.heap)
```

```python
heap = MaxHeap()

heap.heapify([50, 20, 70])
print(heap.heap)
```

## 1-5. 연습 문제
[문제명] 더 맵게
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/42626

```python
import heapq

def solution(scoville, K):
    answer = 0
    heapq.heapify(scoville)  # 기존 리스트를 힙으로 변환
    
    # 가장 작은 값이 K보다 작은 조건에서 반복
    while scoville[0] < K:
        # 음식이 2개 미만으로 남았는데 K를 넘지 못한 경우
        if len(scoville) < 2:
            return -1
        
        # 가장 맵지 않은 두 개를 꺼내서 섞기
        first = heapq.heappop(scoville)
        second = heapq.heappop(scoville)
        
        new_scoville = first + (second * 2)
        heapq.heappush(scoville, new_scoville)
        
        answer += 1
    return answer
```

```python
print(solution(scoville=[1, 2, 3, 9, 10, 12], K=7))
```

# 2. 우선순위 큐
## 2-1. 최소 우선순위 큐 구현 - heapq 모듈 활용

```python
import heapq

class MinPriorityQueue:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def peek(self):
        return self.queue[0][1]  # 우선순위가 가장 높은 요소의 item 확인
    
    def enqueue(self, item, priority):
        # 파이썬 heapq는 최소 힙이므로 작은 숫자가 높은 우선순위라고 가정
        heapq.heappush(self.queue, (priority, item))
    
    def dequeue(self):
        return heapq.heappop(self.queue)[1]  # 우선순위가 가장 높은 요소 삭제 및 item 반환
```

```python
mpq = MinPriorityQueue()

mpq.enqueue("일반 작업", 50)
mpq.enqueue("매우 중요한 작업", 10)
mpq.enqueue("중간 작업", 25)
print(mpq.queue)

mpq.dequeue()
print(mpq.queue)
```

## 2-2. 최대 우선순위 큐 구현 - heapq 모듈 활용

```python
import heapq

class MaxPriorityQueue:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def peek(self):
        return self.queue[0][1]
    
    def enqueue(self, item, priority):
        # 최대 힙 효과를 위해 priority에 -1을 곱하여 삽입
        heapq.heappush(self.queue, (-1 * priority, item))
    
    def dequeue(self):
        return heapq.heappop(self.queue)[1]
```

```python
mpq = MaxPriorityQueue()

mpq.enqueue("일반 작업", 10)
mpq.enqueue("매우 중요한 작업", 50)
mpq.enqueue("중간 작업", 25)
print(mpq.queue)

mpq.dequeue()
print(mpq.queue)
```
