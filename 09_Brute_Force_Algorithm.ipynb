# **1. 선택 정렬**
## 1-1. 선택 정렬 구현

```python
def selection_sort(A):
    n = len(A)                             # 리스트의 크기
    for i in range(n-1):                   # i범위: 0 ~ n-2
        least = i                          # 최소 요소의 인덱스
        for j in range(i+1, n):            # j범위: i+1 ~ n-1
            if A[j] < A[least]:            # A[j]가 더 작으면
                least = j                  # 최소 항목 갱신
        A[i], A[least] = A[least], A[i]	   # A[i]와 A[least] 교환
    return A
```

```python
print(selection_sort(A=[5, 3, 8, 4, 9, 1, 6, 2, 7]))
```

## 1-2. 연습 문제
[문제명] K번째 수
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/42748

```python
def solution(array, commands):
    answer = []
    for cmd in commands:
        i, j, k = cmd
        # 1. 자르기 (인덱스는 0부터 시작하므로 i-1부터 j까지)
        sub_array = array[i-1:j]
        
        # 2. 선택 정렬(Selection Sort) 수행
        n = len(sub_array)
        for step in range(n-1):
            min_idx = step
            for idx in range(step + 1, n):
                if sub_array[idx] < sub_array[min_idx]:
                    min_idx = idx
            sub_array[step], sub_array[min_idx] = sub_array[min_idx], sub_array[step]
        
        # 3. k번째 수 추출 (인덱스는 k-1)
        answer.append(sub_array[k-1])
    return answer
```

```python
print(solution(array=[1, 5, 2, 6, 3, 7, 4], commands=[[2, 5, 3], [4, 4, 1], [1, 7, 3]]))
```

# **2. 순차 탐색**
## 2-1. 순차 탐색 구현

```python
def sequential_search(A, key):      # 순차탐색
    for i in range(len(A)):         # i : 0, 1, ... len(A)-1
        if A[i] == key:             # 탐색 성공하면
            return i                # 인덱스 반환 
    return -1                       # 탐색 실패시 -1 반환 
```

```python
print(sequential_search(A=[5, 3, 8, 4, 2, 7], key=4))
print(sequential_search(A=[5, 3, 8, 4, 2, 7], key=6))
```

## 2-2. 연습 문제
[문제명] 스킬 트리
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/49993

```python
def solution(skill, skill_trees):
    answer = 0
    
    for tree in skill_trees:
        # 선행 스킬에 포함된 문자만 순차적으로 추출
        filtered_skill = ""
        for char in tree:
            if char in skill:
                filtered_skill += char
        
        # 추출된 스킬이 원래 skill의 순서와 일치하는지 확인
        if skill.startswith(filtered_skill):
            answer += 1
    return answer
```

```python
print(solution(skill="CBD", skill_trees=["BACDE", "CBADF", "AECB", "BDA"]))
```

# **3. 완전 탐색**
## 3-1. 완전 탐색 구현

```python
def Knapsack_BF(wgt, val, W):
    n = len(wgt)                  # 전체 물건의 개수
    bestVal = 0                   # 배낭의 최대 가치
    for i in range(2**n):         # 부분 집합의 수: 2^n
        s = [0] * n
        for d in range(n):        # i를 이진수로 변환
            s[d] = i % 2          # 1이 배낭에 넣는 물건 위치를 나타냄
            i = i // 2
        sumVal, sumWgt = 0, 0
        for d in range(n):        
            if s[d] == 1:
               sumWgt += wgt[d]   # 현재 경우의 수의 무게 합
               sumVal += val[d]   # 현재 경우의 수의 가치 합
        if sumWgt <= W:
            if sumVal > bestVal:
                bestVal = sumVal  # 가능한 부분 집합 중 최대 가치 저장
    return bestVal
```

```python
print(Knapsack_BF(wgt=[10, 20, 30, 25, 35], val=[60, 100, 120, 70, 85], W=80))
```

## 3-2. 연습 문제
[문제명] 소수 찾기
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/42839

```python
from itertools import permutations

def is_prime(n):  # 소수 판별 함수
    if n < 2:
        return False
    # 2부터 n의 제곱근까지 나누어 떨어지는지 확인
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def solution(numbers):
    all_numbers = set()
    
    # 모든 길이의 순열 생성
    for i in range(1, len(numbers) + 1):
        # permutations를 통해 i개의 숫자 조각을 뽑는 모든 경우의 수 생성
        perms = permutations(list(numbers), i)
        for p in perms:
            # 튜플 형태의 숫자를 합쳐서 정수로 변환
            num = int(''.join(p))
            all_numbers.add(num)
            
    # 중복을 제거한 숫자들 중 소수의 개수 카운트
    answer = 0
    for n in all_numbers:
        if is_prime(n):
            answer += 1
            
    return answer
```

```python
print(solution(numbers="17"))
print(solution(numbers="011"))
```

# **4. 문자열 매칭**
## 4-1. 문자열 매칭 구현

```python
def string_matching(T, P):               # T는 입력, P는 패턴
    n = len(T)                           # 입력 문자열의 길이
    m = len(P)                           # 패턴 문자열의 길이

    for i in range(n-m+1):               # i: 0, 1, ..., n-m
        j = 0                           
        while j < m and P[j] == T[i+j]:  # 패턴 문자열을 모두 비교
            j += 1
        if j == m:                       # 모든 문자가 일치하면 매칭 성공
            return i                     # 문자열 매칭 성공 시 현재 위치 반환
    return -1                            # 문자열 매칭 실패 시 -1 반환
```

```python
print(string_matching(T="HELLO WORLD", P="LO"))
print(string_matching(T="HELLO WORLD", P="HI"))
```

## 4-2. 연습 문제
[문제명] 문자열 압축
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/60057 

```python
def solution(s):
    # 문자열 길이가 1인 경우 예외 처리
    if len(s) == 1:
        return 1
    
    answer = len(s)
    
    # 단위를 1부터 len(s)//2까지 늘려가며 확인 (억지 기법)
    for step in range(1, len(s) // 2 + 1):
        compressed = ""
        prev = s[0:step]  # 앞에서부터 step만큼의 문자열 추출
        count = 1
        
        # step 크기만큼 증가시키며 이전 문자열과 비교
        for j in range(step, len(s), step):
            # 이전 상태와 동일하다면 압축 횟수 증가
            if prev == s[j:j + step]:
                count += 1
            # 다른 문자열이 나왔다면 (더 이상 압축되지 않는다면)
            else:
                if count >= 2:
                    compressed += str(count) + prev
                else:
                    compressed += prev
                prev = s[j:j + step]  # 상태 초기화
                count = 1
        
        # 마지막에 남은 문자열 처리
        if count >= 2:
            compressed += str(count) + prev
        else:
            compressed += prev
        
        # 만들어진 압축 문자열 중 가장 짧은 길이 선택
        answer = min(answer, len(compressed))
        
    return answer
```

```python
print(solution(s="aabbaccc"))
print(solution(s="ababcdcdababcdcd"))
print(solution(s="abcabcdede"))
print(solution(s="abcabcabcabcdededededede"))
print(solution(s="xababcdcdababcdcd"))
```
