# **1. 일반 트리**
## 1-1. 일반 트리 구현

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.parent = None
        self.children = []
```

```python
class Tree:
    def __init__(self):
        self._root = None
        self._size = 0

    def is_empty(self):
        return self._size == 0

    def size(self):
        return self._size

    def root(self):
        return self._root
            
    def parent(self, node):
        return node.parent

    def children(self, node):
        return node.children

    def add_root(self, value):
        if self._root is not None:
            print("Root already exists")
            return

        self._root = Node(value)
        self._size = 1
        return self._root

    def add_child(self, node, value):
        new_node = Node(value)
        new_node.parent = node
        node.children.append(new_node)

        self._size += 1
        return new_node

    def remove(self, node):
        if node.children:
            print("Cannot delete a node that has children")
            return

        if node is self._root:
            self._root = None
            self._size = 0
            return node
        
        parent = node.parent
        if parent:
            parent.children.remove(node)
        
        self._size -= 1
        node.parent = None
        return node

    def height(self):
        return self._height(self._root)

    def _height(self, node):
        if node is None:
            return -1
        return 1 + max(self._height(child) for child in node.children)
```

```python
tree = Tree()

# 루트 생성
A = tree.add_root("A")

B = tree.add_child(A, "B")
C = tree.add_child(A, "C")

D = tree.add_child(B, "D")
E = tree.add_child(B, "E")
```

## 1-2. 연습 문제
[문제명] 다단계 칫솔 판매
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/77486

```python
def solution(enroll, referral, seller, amount):
    # 부모 관계를 나타내는 딕셔너리 생성
    parent = {}
    for e, r in zip(enroll, referral):
        parent[e] = r
    
    # 각 판매원의 이름을 키로, 총 수익을 값으로 하는 딕셔너리 생성
    total = {name: 0 for name in enroll}
    
    for s, a in zip(seller, amount):
        money = a * 100  # 개당 100원
        curr_seller = s
        
        # 부모를 타고 올라가며 수익 배분
        while curr_seller != "-" and money > 0:
            tax = money // 10  # 나를 추천한 사람에게 줄 10% (원 단위 절사)
            total[curr_seller] += (money - tax)  # 내가 가질 90%
            
            # 다음 배분 대상과 금액 갱신
            curr_seller = parent[curr_seller]
            money = tax
            
            # 배분할 금액이 0원이 되면 더 이상 올라갈 필요 없음
            if money == 0:
                break
                
    # enroll에 담긴 순서대로 결과 반환
    return [total[name] for name in enroll]
```

```python
print(solution(enroll=["john", "mary", "edward", "sam", "emily", "jaimie", "tod", "young"], referral=["-", "-", "mary", "edward", "mary", "mary", "jaimie", "edward"], seller=["young", "john", "tod", "emily", "mary"], amount=[12, 4, 2, 5, 10]))
print(solution(enroll=["john", "mary", "edward", "sam", "emily", "jaimie", "tod", "young"], referral=["-", "-", "mary", "edward", "mary", "mary", "jaimie", "edward"], seller=["sam", "emily", "jaimie", "edward"], amount=[2, 3, 5, 4]))
```

# **2. 이진 트리**
## 2-1. 이진 트리 구현

```python
class BTNode:
    def __init__(self, value):
        self.value = value
        self.parent = None
        self.left = None
        self.right = None
```

```python
from collections import deque

class BinaryTree:
    def __init__(self):
        self._root = None
        self._size = 0

    def is_empty(self):
        return self._size == 0

    def size(self):
        return self._size

    def root(self):
        return self._root

    def parent(self, node):
        return node.parent

    def left(self, node):
        return node.left

    def right(self, node):
        return node.right

    def add_root(self, value):
        if self._root is not None:
            print("Root already exists")
            return
        
        self._root = BTNode(value)
        self._size = 1
        return self._root

    def add_left(self, node, value):
        if node.left is not None:
            print("Left child already exists")
            return
        
        new_node = BTNode(value)
        new_node.parent = node
        node.left = new_node
        
        self._size += 1
        return new_node

    def add_right(self, node, value):
        if node.right is not None:
            print("Right child already exists")
            return
        
        new_node = BTNode(value)
        new_node.parent = node
        node.right = new_node

        self._size += 1
        return new_node

    def remove(self, node):
        if node.left is not None or node.right is not None:
            print("Cannot delete a node that has children")
            return

        if node is self._root:
            self._root = None
            self._size = 0
            return node

        parent = node.parent
        if parent.left is node:
            parent.left = None
        else:
            parent.right = None

        self._size -= 1
        node.parent = None
        return node

    def height(self):
        return self._height(self._root)

    def _height(self, node):
        if node is None:
            return -1
        return 1 + max(self._height(node.left), self._height(node.right))
    
    def preorder(self, node):
        if node:
            print(node.value)
            self.preorder(node.left)
            self.preorder(node.right)

    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.value)
            self.inorder(node.right)

    def postorder(self, node):
        if node:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.value)

    def level_order(self):
        if self._root is None:
            return
        
        queue = deque([self._root])
        while queue:
            node = queue.popleft()
            print(node.value)
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
```

```python
tree = BinaryTree()

# 루트 생성
A = tree.add_root("A")

B = tree.add_left(A, "B")
C = tree.add_right(A, "C")

D = tree.add_left(B, "D")
E = tree.add_right(B, "E")
```

## 2-2. 연습 문제
[문제명] 길 찾기 게임
</br>[링크] https://school.programmers.co.kr/learn/courses/30/lessons/42892

```python
import sys
sys.setrecursionlimit(10**6)  # 재귀 깊이 초과 방지

class BTNode:
    def __init__(self, value):
        self.value = value
        self.parent = None
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self._root = None
        self._size = 0

    def add_root(self, value):
        if self._root is not None:
            print("Root already exists")
            return
        
        self._root = BTNode(value)
        self._size = 1
        return self._root

    def add_left(self, node, value):
        if node.left is not None:
            print("Left child already exists")
            return
        
        new_node = BTNode(value)
        new_node.parent = node
        node.left = new_node

        self._size += 1
        return new_node

    def add_right(self, node, value):
        if node.right is not None:
            print("Right child already exists")
            return
        
        new_node = BTNode(value)
        new_node.parent = node
        node.right = new_node
        
        self._size += 1
        return new_node
    
    def preorder(self):
        result = []
        self._preorder(self._root, result)
        return result
    
    def _preorder(self, node, result):
        if node:
            result.append(node.value)
            self._preorder(node.left, result)
            self._preorder(node.right, result)

    def postorder(self):
        result = []
        self._postorder(self._root, result)
        return result
    
    def _postorder(self, node, result):
        if node:
            self._postorder(node.left, result)
            self._postorder(node.right, result)
            result.append(node.value)
```

```python
def solution(nodeinfo):
    tree = BinaryTree()

    # (x, y, node_number)
    nodes = []
    for i, (x, y) in enumerate(nodeinfo, start=1):
        nodes.append((x, y, i))

    # y 내림차순, x 오름차순 정렬
    nodes.sort(key=lambda v: (-v[1], v[0]))

    # 첫 번째 노드를 루트 노드로 지정
    tree.add_root(nodes[0])

    # 이후 노드들을 x 기준 BST처럼 삽입
    for value in nodes[1:]:
        x, y, idx = value
        current = tree._root

        while True:
            cur_x, cur_y, cur_idx = current.value

            if x < cur_x:
                if current.left is None:
                    tree.add_left(current, value)
                    break
                current = current.left
            else:
                if current.right is None:
                    tree.add_right(current, value)
                    break
                current = current.right

    preorder_result = tree.preorder()
    postorder_result = tree.postorder()

    preorder_result = [value[2] for value in preorder_result]
    postorder_result = [value[2] for value in postorder_result]
    return [preorder_result, postorder_result]
```

```python
print(solution(nodeinfo=[[5,3],[11,5],[13,3],[3,5],[6,1],[1,3],[8,6],[7,2],[2,2]]))
```

# **3. 이진 탐색 트리**
## 3-1. 이진 탐색 트리 구현

```python
class BSTNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
```

```python
class BinarySearchTree:
    def __init__(self):
        self._root = None
        self._size = 0

    def is_empty(self):
        return self._size == 0

    def size(self):
        return self._size

    def root(self):
        return self._root

    def left(self, node):
        return node.left

    def right(self, node):
        return node.right

    def height(self):
        return self._height(self._root)

    def _height(self, node):
        if node is None:
            return -1
        return 1 + max(self._height(node.left), self._height(node.right))

    def inorder(self):
        self._inorder(self._root)

    def _inorder(self, node):
        if node:
            self._inorder(node.left)
            print(node.key)
            self._inorder(node.right)

    def search(self, key):
        return self._search(self._root, key)

    def _search(self, node, key):
        if node is None:  # node가 None이면 공백 트리이므로 None 반환
            return None
        elif key < node.key:  # 탐색 키가 node의 key보다 작으면 왼쪽 서브 트리 탐색
            return self._search(node.left, key)
        elif key > node.key:  # 탐색 키가 node의 key보다 크면 오른쪽 서브 트리 탐색
            return self._search(node.right, key)
        else:  # 탐색 키가 node의 key와 일치하면 해당 node 반환
            return node

    def insert(self, key):
        if not self.search(key):  # 중복이 아닐 때만
            self._root = self._insert(self._root, key)
            self._size += 1

    def _insert(self, node, key):
        if node is None:  # node가 None이면 해당 위치에 key 값을 가지는 노드 삽입
            return BSTNode(key)
        elif key < node.key:  # 삽입 키가 node의 key보다 작으면 왼쪽 서브 트리에서 삽입
            node.left = self._insert(node.left, key)
        elif key > node.key:  # 삽입 키가 node의 key보다 크면 오른쪽 서브 트리에서 삽입
            node.right = self._insert(node.right, key)
        else:  # 삽입 키가 node의 key와 일치하면 해당 node 반환(동일 키 미허용)
            return node

    def delete(self, key):
        if self.search(key):  # 존재할 때만
            self._root = self._delete(self._root, key)
            self._size -= 1

    def _delete(self, node, key):
        if node is None:  # node가 None이면 공백 트리이므로 None 반환
            return None
        elif key < node.key:  # 삭제 키가 node의 key보다 작으면 왼쪽 서브 트리에서 삭제
            node.left = self._delete(node.left, key)
        elif key > node.key:  # 삭제 키가 node의 key보다 크면 오른쪽 서브 트리에서 삭제
            node.right = self._delete(node.right, key)
        else:  # 삭제 키가 node의 key와 일치하는 경우
            # 경우 1 & 2: 자식이 없거나 하나인 경우
            if node.left is None:
                return node.right  # 삭제할 노드를 오른쪽 자식으로 대체
            elif node.right is None:
                return node.left  # 삭제할 노드를 왼쪽 자식으로 대체

            # 경우 3: 자식이 둘인 경우
            temp = node.right
            while temp.left is not None:  # 오른쪽 서브 트리에서 가장 작은 노드 탐색
                temp = temp.left

            node.key = temp.key  # 삭제할 노드를 오른쪽 서브 트리의 가장 작은 노드로 대체
            node.right = self._delete(node.right, temp.key)  # 오른쪽 서브 트리에서 대체 노드 삭제
        return node
```
